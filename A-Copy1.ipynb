```python
import string
import nltk
from nltk.corpus import stopwords
from collections import Counter
import matplotlib.pyplot as plt
```

```python
text = open('readme.txt',encoding='utf-8').read()
print(text)
```

```python
#convert the letters into lower case
lower_case = text.lower()
print(lower_case)
```

```python
#removing puncutations
print(string.punctuation)
```

```python
cleaned_txt = lower_case.translate(str.maketrans('','',string.punctuation))
cleaned_txt

#https://www.w3schools.com/python/ref_string_maketrans.asp
```

### Tokenization and stop word

```python
tokenized_words = cleaned_txt.split()
tokenized_words
```

```python
#stop word dont add any meaning to the sentence , like in the text fle...."i" doesnot add any emotion
stop_words = ["i", "me", "my", "myself", "we", "our", "ours", "ourselves", "you", "your", "yours", "yourself",
              "yourselves", "he", "him", "his", "himself", "she", "her", "hers", "herself", "it", "its", "itself",
              "they", "them", "their", "theirs", "themselves", "what", "which", "who", "whom", "this", "that", "these",
              "those", "am", "is", "are", "was", "were", "be", "been", "being", "have", "has", "had", "having", "do",
              "does", "did", "doing", "a", "an", "the", "and", "but", "if", "or", "because", "as", "until", "while",
              "of", "at", "by", "for", "with", "about", "against", "between", "into", "through", "during", "before",
              "after", "above", "below", "to", "from", "up", "down", "in", "out", "on", "off", "over", "under", "again",
              "further", "then", "once", "here", "there", "when", "where", "why", "how", "all", "any", "both", "each",
              "few", "more", "most", "other", "some", "such", "no", "nor", "not", "only", "own", "same", "so", "than",
              "too", "very", "s", "t", "can", "will", "just", "don", "should", "now"]
```

```python
final_word=[]
for word in tokenized_words:
    if(word not in stop_words):
        final_word.append(word)
final_word
```

### Emotion Algorithm

```python
print(final_word)
```

```python
# if word is present ----> add the emotion to the emotion list
# finally count each_eomotion int the emotion list
```

```python
emotion_list = []
with open('emotion.txt','r') as file:
    for line in file:
        clear_line = line.replace("\n",'').replace(',','').replace("'",'').strip()
        #print(clear_line)
        word,emotion=clear_line.split(':')
        #print('Word: '+word+" "+'Emotion: '+emotion)
        
        if word in final_word:
            emotion_list.append(emotion)
print(emotion_list)
w=Counter(emotion_list)
print(w)
```

```python
plt.bar(w.keys(),w.values())
plt.savefig('Graph.png')
plt.show()
```
